## Investigation Overview



The overall goal of my presentation was to analyze which variables had an effect on the original loan payment for each borrower. I also sought to analyze the variables that had an effect on the credit grade for each borrower, pertaining to the Proper Loan dataset.

## Dataset Overview and Executive Summary


The Prosper Loan Dataset contains 113,937 loans with 81 variables on each loan. The analyzing variables includes the loan amount, borrower rate, current loan status, employment status, credit grade, and more.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
pld_df = pd.read_csv('prosperLoanData.csv')
````

````python
#Function for initializing plot details
def plot_label(title, x_label, y_label):
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    
````

## Loan Status vs Original Loan Amount



This visualization is a box plot showing the completed and defaulted loan statuses, compared to the original loan amount. This visualization gives us insight, showing that a completed loan status contains many values of higher loan amounts than for a defaulted loan status.



````python
loanCondition = (pld_df['LoanStatus'] == 'Completed') | (pld_df['LoanStatus'] == 'Defaulted')
new_loanstatus_df = pld_df[loanCondition]
sns.boxplot(data=new_loanstatus_df, x='LoanStatus', y='LoanOriginalAmount');
plot_label('Loan Status vs Original Loan Amount', 'Loan Status', 'Loan Amount')
````

## Credit Grade vs Monthly Loan Payment

This visualization is a violin plot that shows the relationship between credit grade and monthly loan payment. This visualization reveals to us that the highest credit grades contained that highest outliers and distributions for monthly loan payments.

````python
sns.violinplot(data=pld_df, x = "CreditGrade", y="MonthlyLoanPayment");
plot_label('Credit Grade vs Monthly Loan Payment', 'Credit Grade', 'Monthly Loan Payment')
````

## Original Loan Amount, Loan Status, and Employment Status

> This visualization is a mult-variate violin plot, showing the relationship between the employment status, loan status, and original loan amount. This visualization reveals to us that employed borrowers with a completed loan status contains the highest loan amounts. It also shows that unemployed borrowers with a defaulted loan status contained the lowest loan amounts.

````python
ordered_rating = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR', 'NC']
plt.figure(figsize=[12,8])
sns.violinplot(data = new_loanstatus_df, x = 'EmploymentStatus', y='LoanOriginalAmount', hue='LoanStatus', rating = ordered_rating);
plot_label('Visualization of Employment Status, Loan Amount, and Loan Status', 'Employment Status ', 'Original Loan Amount')
````
